```python
def convert_keys_to_int(d):
    new_dict = {}
    for k, v in d.items():
        new_key = int(k) if k.isdigit() else k
        if isinstance(v, dict):
            new_dict[new_key] = convert_keys_to_int(v)
        else:
            new_dict[new_key] = v
    return new_dict
```

```python
def get_kmeans(embeddings):
    # List of k values
    k_values = [5, 10, 15, 20, 25]
    # Dictionary to store cluster labels for each k
    cluster_results = {}
    
    # Perform K-means clustering for each k
    for k in k_values:
        kmeans = KMeans(n_clusters=k, random_state=42)
        cluster_labels = kmeans.fit_predict(embeddings)
        cluster_results[k] = cluster_labels
    
    # # Print the dictionary keys to verify the results
    # print(cluster_results.keys())
    
    # # Optionally, print the first few labels for each k
    # for k, labels in cluster_results.items():
    #     print(f"Cluster labels for k={k}: {labels[:10]}")
    return cluster_results
```

```python
def get_silhouette_score(embeddings, predict_kmeans):
    silhouette_scores = []
    for k in k_values:
        silhouette_scores.append(silhouette_score(embeddings, predict_kmeans[k]))
    
    return silhouette_scores
```

```python
def generate_silhouette_score(ctype):
    text_graph_src = {}
    text_gnn_src_list = []
    text_hop_gnn_src_list = []
    node2vec_gnn_src_list = []
    deepwalk_gnn_src_list = []
    text_src = []
    text_hop_src = []
    node2vec_src=[]
    deepwalk_src=[]
    # if(ctype=="attack_graph" or ctype=="attack_text"):
    #     base_kmeans = get_kmeans(base[:attack_size])
    # elif(ctype=="weak_graph" or ctype=="weak_text"):
    #     base_kmeans = get_kmeans(base[attack_size:])
    # get_kmeans(gnn_deepwalk_e[-1])
    for i,model in enumerate(models):
        print("Model: ", model)
        text_gnn_src = []
        hop_gnn_src = []
        node2vec_gnn_src = []
        deepwalk_gnn_src = []
        for j in range(len(gnn_models)):
            print("\tGNN: ", gnn_models[j])
            # gnn_hop_kmeans.append(gnn_hop)
            # gnn_node2vec_kmeans.append(gnn_node2vec)
            # gnn_deepwalk_kmeans.append(gnn_deepwalk)
            if(ctype=="attack_graph" or ctype=="attack_text"):
                if(dual==""):
                    gnn_text_emb = gnn_embeddings["text"][i][j][:attack_size]
                # gnn_hop_emb=gnn_embeddings["hop_text"][i][j][:attack_size]
                gnn_node2vec_emb = gnn_embeddings["node2vec"][i][j][:attack_size]
                gnn_deepwalk_emb = gnn_embeddings["deepwalk"][i][j][:attack_size]
                
            elif(ctype=="weak_graph" or ctype=="weak_text"):
                if(dual==""):
                    gnn_text_emb = gnn_embeddings["text"][i][j][attack_size:]
                # gnn_hop_emb=gnn_embeddings["hop_text"][i][j][attack_size:]
                gnn_node2vec_emb=gnn_embeddings["node2vec"][i][j][attack_size:]
                gnn_deepwalk_emb=gnn_embeddings["deepwalk"][i][j][attack_size:]
            else:
                if(dual==""):
                    gnn_text_emb = gnn_embeddings["text"][i][j]
                # gnn_hop_emb=gnn_embeddings["hop_text"][i][j]
                gnn_node2vec_emb=gnn_embeddings["node2vec"][i][j]
                gnn_deepwalk_emb=gnn_embeddings["deepwalk"][i][j]
            if(dual==""):
                sill_score = get_silhouette_score(gnn_text_emb,get_kmeans(gnn_text_emb))
                text_gnn_src.append(sill_score)
            # sill_score = get_silhouette_score(gnn_hop_emb,get_kmeans(gnn_hop_emb))
            # hop_gnn_src.append(sill_score)
            sill_score = get_silhouette_score(gnn_node2vec_emb,get_kmeans(gnn_node2vec_emb))
            node2vec_gnn_src.append(sill_score)
            sill_score = get_silhouette_score(gnn_deepwalk_emb,get_kmeans(gnn_deepwalk_emb))
            deepwalk_gnn_src.append(sill_score)
        if(dual==""):
            text_gnn_src_list.append(text_gnn_src)
        # text_hop_gnn_src_list.append(hop_gnn_src)
        node2vec_gnn_src_list.append(node2vec_gnn_src)
        deepwalk_gnn_src_list.append(deepwalk_gnn_src)
        if(ctype=="attack_graph" or ctype=="attack_text"):
            text_emb = text_hop_embeddings["text"][i][:attack_size]
            # hop_emb= text_hop_embeddings["hop_text"][i][:attack_size]

        elif(ctype=="weak_graph" or ctype=="weak_text"):
            text_emb =text_hop_embeddings["text"][i][attack_size:]
            # hop_emb= text_hop_embeddings["hop_text"][i][attack_size:]
        else:
            text_emb =text_hop_embeddings["text"][i]
            # hop_emb= text_hop_embeddings["hop_text"][i]
        sill_score = get_silhouette_score(text_emb,get_kmeans(text_emb))
        text_src.append(sill_score)
        # sill_score = get_silhouette_score(hop_emb,get_kmeans(hop_emb))
        # text_hop_src.append(sill_score)
    if(ctype=="attack_graph" or ctype=="attack_text"):
        node2vec_emb = graph_embeddings["node2vec"][:attack_size]
        deepwalk_emb= graph_embeddings["deepwalk"][:attack_size]

    elif(ctype=="weak_graph" or ctype=="weak_text"):
        node2vec_emb = graph_embeddings["node2vec"][attack_size:]
        deepwalk_emb = graph_embeddings["deepwalk"][attack_size:]
    else:
        node2vec_emb = graph_embeddings["node2vec"]
        deepwalk_emb = graph_embeddings["deepwalk"]
    sill_score = get_silhouette_score(node2vec_emb,get_kmeans(node2vec_emb))
    node2vec_src=(sill_score)
    sill_score = get_silhouette_score(deepwalk_emb,get_kmeans(deepwalk_emb))
    deepwalk_src=(sill_score)
    if(dual==""):
        text_graph_src["text_gnn"]=text_gnn_src_list
    # text_graph_src["text_hop_gnn"]=text_hop_gnn_src_list
    text_graph_src["node2vec_gnn"]=node2vec_gnn_src_list
    text_graph_src["deepwalk_gnn"]=deepwalk_gnn_src_list
    text_graph_src["text"]=text_src
    # text_graph_src["text_hop"]=text_hop_src
    text_graph_src["node2vec"]=node2vec_src
    text_graph_src["deepwalk"]=deepwalk_src
    return text_graph_src
```
