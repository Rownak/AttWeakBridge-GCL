```python
def convert_keys_to_int(d):
    new_dict = {}
    for k, v in d.items():
        new_key = int(k) if k.isdigit() else k
        if isinstance(v, dict):
            new_dict[new_key] = convert_keys_to_int(v)
        else:
            new_dict[new_key] = v
    return new_dict
```

```python
import random

def generate_rand(start, end, k):
    return random.sample(range(start, end), k)
```

```python

def calc_rand_intersect2(bron_attack_cwe, sampling_time, isAttack):
    # Loop over each k from 10 to 1136
    average_hit_at_k = {}
    average_hit_at_k_avg = {}
    if(isAttack):
        size = attack_size
    else:
        size = weakness_size
    for k in range(5, 80,5):
        hit_at_k = []
        hit_at_k_avg = []
        row_add = 0

        for i in range(size):
            # Get the indices of the top k values in the row for both matrices
            node_id =i
            if(not isAttack):
                node_id = i+attack_size
            if(node_id not in bron_attack_cwe.keys()):
                continue
            # if(len(bron_attack_cwe[i])<k):
            #     continue
            top_bron = bron_attack_cwe[node_id]
            hit_count=0
            for _ in range(sampling_time):
                if(isAttack):
                    top_k = generate_rand(attack_weak_range['cwe'][0], attack_weak_range['cwe'][1], k)
                else:
                    top_k = generate_rand(attack_weak_range['attack'][0], attack_weak_range['attack'][1], k)
                # Calculate the hit@k (number of common indices in top k)
                #hit_count = len(set(top_bron) & set(top_k))
                hit_count += len(set(top_bron) & set(top_k))

            hit_count = hit_count/float(sampling_time)
            # Append the result to the list
            hit_at_k_avg.append(hit_count/float(min(k,len(top_bron))))
            hit_at_k.append(hit_count)
        # Calculate the average hit@k for the current k
        #print(hit_at_k)
        average_hit_at_k[k] = np.mean(hit_at_k)
        average_hit_at_k_avg[k] = np.mean(hit_at_k)
    return average_hit_at_k, average_hit_at_k_avg
```

```python

def calc_intersect(bron_attack_cwe, cosine_matrix, isAttack):
    # Loop over each k from 10 to 1136
    average_hit_at_k = {}
    if(isAttack):
        size = attack_size
    else:
        size = weakness_size
    for k in range(5, 80,5):
        hit_at_k = []
        for i in range(size):
            # Get the indices of the top k values in the row for both matrices
            node_id =i
            if(not isAttack):
                node_id = i+attack_size
            if(node_id not in bron_attack_cwe.keys()):
                continue
            # if(len(bron_attack_cwe[i])<k):
            #     continue
            top_bron = bron_attack_cwe[node_id]
            top_k = np.argsort(cosine_matrix[i])[-k:]
            if(isAttack):
                top_k = [x+attack_size for x in top_k]
            
            # Calculate the hit@k (number of common indices in top k)
            #hit_count = len(set(top_bron) & set(top_k))
            if(isAverage):
                hit_count = len(set(top_bron) & set(top_k))/float(min(k,len(top_bron)))
            else:
                hit_count = len(set(top_bron) & set(top_k))
            
            # Append the result to the list
            hit_at_k.append(hit_count)
        # Calculate the average hit@k for the current k
        # print(hit_at_k)
        average_hit_at_k[k] = np.mean(hit_at_k)
    return average_hit_at_k
```

```python

def plot_hit_k_graph_gnn(bron_attack_cwe,matrix1, matrix2, matrix3, isAttack):
    
    # Initialize a dictionary to store average hit@k for each k
    markers = ['o', 's', '+', 'v', '<', '>']
    colors = list(mcolors.TABLEAU_COLORS)
    colors.pop(3)
    plt.figure(figsize=(5, 5))
    # Loop over all the models
    #models = ["pretrained_SecBert", "SecBert", "pretrained_SecureBert",  "SecureBert", "pretrained_Gpt2","Gpt2"]
    #models = ["Gpt2"]
    if(isAttack):
        size = attack_size
    else:
        size = weakness_size
    all_hits = []
    for i in range(0,len(models),1):
        # Define the cosine similarity matrices
        model = models[i]
        if(isAttack):
            cosine_matrix2 = matrix2[i][:203,203:]
            cosine_matrix3 = matrix3[i][:203,203:]
        else:
            cosine_matrix2 = matrix2[i][203:,:203]
            cosine_matrix3 = matrix3[i][203:,:203]
        for j in range(len(gnn_models)):
            if(isAttack):
                cosine_matrix1 = matrix1[i][j][:203,203:]
            else:
                cosine_matrix1 = matrix1[i][j][203:,:203]
            average_hit_at_k1 = calc_intersect(bron_attack_cwe, cosine_matrix1, isAttack)
            ks = list(average_hit_at_k1.keys())
            hits1 = list(average_hit_at_k1.values())
            mrk = markers[i]
            if(j==2 or j==5):
                if(i==0):
                    mrk = '*'
                else:
                    mrk = 'p'
            plt.plot(ks, hits1, label=model+" "+gnn_models[j], marker=mrk, color = colors[j+2])
            # print("attack calculated: ", count)
        # Plot the results
        average_hit_at_k2 = calc_intersect(bron_attack_cwe, cosine_matrix2, isAttack)
        average_hit_at_k3 = calc_intersect(bron_attack_cwe, cosine_matrix3, isAttack)
        if(isAttack):
            average_hit_at_k4 = rand_hit_at_k_attack
        else:
            average_hit_at_k4 = rand_hit_at_k_weak
        ks = list(average_hit_at_k2.keys())
        hits2 = list(average_hit_at_k2.values())
        hits3 = list(average_hit_at_k3.values())
        hits4 = list(average_hit_at_k4.values())
        plt.plot(ks, hits2, label=model+" Text emb", marker='.', color = colors[i])
        plt.plot(ks, hits3, label=model+" Hop emb", marker='v', color = colors[i])
        plt.plot(ks, hits4, label=model+" Rand Pick", marker='x', color = 'tab:red')
        print(hits3)
    plt.xlabel('K')
    plt.ylabel('Average Hit@k')
    plt.xticks(ks)
    plt.grid(True)
    plt.legend(bbox_to_anchor=(1.1, 1.05))
    #plt.title('GNN Semi Supervised - Text - Hop Embeddings')
    if(isAttack):
        plt.title('Bron-GNN Hit@k Attack to Weakness')
    else:
        plt.title('Bron-GNN Hit@k Weakness to Attack')

    if(isAttack):
        if(isAverage):
            title = "rand_{}_avg_bron_hit@k_gnn_sample_{}_attack_weak.png".format(sample,rand)
        else:
            title = "rand_{}_bron_hit@k_gnn_sample_{}_attack_weak.png".format(sample,rand)
    else:
        if(isAverage):
            title = "rand_{}_avg_bron_hit@k_gnn_sample_{}_weak_attack.png".format(sample,rand)
        else:
            title = "rand_{}_bron_hit@k_gnn_sample_{}_weak_attack.png".format(sample,rand)
            
    plt.savefig(plot_dir+title, dpi=300, bbox_inches='tight')
    plt.show()
```
