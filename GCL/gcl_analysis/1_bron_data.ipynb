```python
import csv
from collections import defaultdict
import json
import sys
#base_dir = os.environ['AWEB_DIR']
sys.path.append("../../")
import config
```

```python
# Initialize the defaultdict
cwe_dict = defaultdict(list)
data_dir = config.DATA_DIR
# Open and read the CSV file
with open(data_dir+'compressed_e.csv', mode='r') as file:  # Replace 'yourfile.csv' with your actual file path
    csv_reader = csv.DictReader(file)
    for row in csv_reader:
        # Get the id and CWEs from the row
        id_value = row['id']
        cwe_values = row['CWEs']
        
        # Split the CWEs by dash and add to the list in the dictionary
        if cwe_values != 'N/A':  # Handle cases where CWEs might be 'N/A'
            cwe_list = cwe_values.split('-')
            cwe_dict[id_value].extend(cwe_list)

# # Print the resulting dictionary
# for key, value in cwe_dict.items():
#     print(f"{key}: {value}")
```

```python
with open(data_dir+"doc_id_to_emb_id.json") as f:
    doc_id_to_emb_id = json.load(f)
with open(data_dir+"emb_id_to_doc_id.json") as f:
    emb_id_to_doc_id = json.load(f)
```

```python
bron_attack_cwe = defaultdict(list)
for attack_key in cwe_dict:
    if(attack_key not in doc_id_to_emb_id.keys()):
        print("attack not found:", attack_key)
        continue
    attack_pos = doc_id_to_emb_id[attack_key]
    for cwe_key in cwe_dict[attack_key]:
        if(cwe_key not in doc_id_to_emb_id.keys()):
            print("cwe not found:", cwe_key)
            continue
        bron_attack_cwe[attack_pos].append(doc_id_to_emb_id[cwe_key])
    print(len(bron_attack_cwe[attack_pos]))
    print(bron_attack_cwe[attack_pos])
    
```

```python
from collections import defaultdict
bron_cwe_attack = defaultdict(list)
for key in bron_attack_cwe:
    for weak in bron_attack_cwe[key]:
        bron_cwe_attack[weak].append(key)
```
